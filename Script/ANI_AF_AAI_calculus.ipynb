```python
import pandas as pd
```

```python
# query is always my org 

def load_data(hit_path,feature_table):
    df_hist= pd.read_csv(hit_path)
    df_feature=pd.read_csv(feature_table,sep="\t")
    return df_hist,df_feature
    
```

```python
hits_table , feature_table = load_data("SR5SKX3F114-my_org_lyticus.csv","myorg_feature_table.txt")
hits_table
```

```python
hits_table.columns
```

```python
feature_table.columns
```

```python
feature_table["# feature"].unique()
```

```python
feature_table=feature_table[feature_table["# feature"]=="CDS"]
feature_table['product_accession'].unique()
```

```python
hits_table["cds_id"]=pd.Series(map(lambda x: x.split("_cds_")[1],hits_table["query acc.ver"]))
hits_table["cds_id"]=pd.Series(map(lambda x: x.rsplit("_",1)[0] if (x[0]=="W") else None ,hits_table["cds_id"]))
hits_table['cds_id'].unique()
```

```python
hits_table
```

```python
hits_table=hits_table[~hits_table["cds_id"].isna()]
```

## Join table vertical search

```python
feature_table=feature_table.rename(columns={"product_accession": "cds_id"})
```

```python
annotated_hits=pd.merge(hits_table,feature_table, on='cds_id',how ='inner')
```

```python
annotated_hits
```

```python
annotated_hits_filtered = annotated_hits[["cds_id","query acc.ver","subject acc.ver","% identity","alignment length","mismatches","gap opens","q. start","q. end","s. start","s. end","evalue","bit score","start","end","strand","non-redundant_refseq","name","feature_interval_length","product_length"]]
```
